```python
#implementation of linear regression from scratch
#push to github
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[2,4,6,8,10]
#y=2x is the correct line
#y=mx+c

def compute_error(m,b,x,y):
    total_error =0
    n= len(x)
    for i in range(n):
        y_pred=m*x[i]+b
        total_error+=(y[i]-y_pred)**2
    return total_error/n

#initiall we start with m,b=0,0
#loss function l=1/n sum(y[actutal]-y[pred])**2
#L=1/n*sum(y-m*x+b)**2
#dl/dm= -2/n*sum(x*(y-(mx+b))
#dl/dm = -2/n*sum(y-(mx*b))
#m_new=m-L*dl/dm
#b_new=b-L*dl/db


def gradient_descent(m_now,b_now,x,y,L):
    m_grad=0
    b_grad=0
    n=len(x)
    for i in range(n):
        y_pred=m_now*x[i]+b_now
        m_grad+=-(2/n)*x[i]*(y[i]-y_pred)
        b_grad+=-(2/n)*(y[i]-y_pred)
    m=m_now-L*m_grad
    b=b_now-L*b_grad
    return m,b

l=0.01
m=0
b=0
x=[1,2,3,4,5]
y=[2,4,6,8,10]
epochs=600
for i in range(epochs):
    m,b=gradient_descent(m,b,x,y,l)
    
    if i%50==0:
        error=compute_error(m,b,x,y)
        print(f"epoch {i}:m={m},b={b},error={error}")
print("m",m)
print("b",b)


plt.scatter(x,y)
reg_line = [m * xi + b for xi in x]
plt.plot(x,reg_line,color='red')
plt.show()
```
