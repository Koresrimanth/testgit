```python
#fixing the outliers in the bill_amount using cqpping
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Define the Hybrid Capping + Log Function
def clean_billing_data(df, col):
    # Calculate IQR for the Upper Cap
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    # Define our "Hybrid" limits
    lower_limit = 0            # Set to 0 because negative debt is 0 risk
    upper_limit = Q3 + 1.5 * IQR # Standard IQR ceiling for outliers
    
    # Step A: Capping (Clip the extremes)
    df[col] = df[col].clip(lower=lower_limit, upper=upper_limit)
    
    # Step B: Log Transformation (Squash the distribution into a bell curve)
    # np.log1p handles the 0 values by calculating log(1 + x)
    df[col] = np.log1p(df[col])

# 2. Apply to all 6 Billing Columns
bill_cols = [f'BILL_AMT{i}' for i in range(1, 7)]

for col in bill_cols:
    clean_billing_data(df, col)

# 3. Visualize the Results
plt.figure(figsize=(15, 10))
for i, col in enumerate(bill_cols, 1):
    plt.subplot(2, 3, i)
    sns.histplot(df[col], kde=True, color='teal')
    plt.title(f'Cleaned & Logged {col}')
plt.tight_layout()
plt.show()

```

```python
#using different kinds of models
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import (RandomForestClassifier, GradientBoostingClassifier,
                               AdaBoostClassifier, ExtraTreesClassifier)
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
from sklearn.metrics import (classification_report, confusion_matrix, 
                              roc_auc_score, roc_curve, auc,
                              precision_recall_curve, f1_score,
                              accuracy_score, precision_score, recall_score,
                              matthews_corrcoef, cohen_kappa_score)
```

```python
models = {
    'Decision Tree': DecisionTreeClassifier(random_state=42, max_depth=10),
    'Random Forest': RandomForestClassifier(random_state=42, n_estimators=100, n_jobs=-1),
    'Gradient Boosting': GradientBoostingClassifier(random_state=42, n_estimators=100),
    'XGBoost': XGBClassifier(random_state=42, n_estimators=100, eval_metric='logloss'),
    'LightGBM': LGBMClassifier(random_state=42, n_estimators=100, verbose=-1),
}
```

```python
def evaluate_model(model, X_train, y_train, X_test, y_test, model_name):
    """Comprehensive model evaluation"""
    # Train model
    model.fit(X_train, y_train)
    
    # Predictions
    y_pred = model.predict(X_test)
    y_pred_proba = model.predict_proba(X_test)[:, 1]
    
    # Calculate metrics
    metrics = {
        'Model': model_name,
        'Precision': precision_score(y_test, y_pred),
        'Recall': recall_score(y_test, y_pred),
        'F1-Score': f1_score(y_test, y_pred),
        'ROC-AUC': roc_auc_score(y_test, y_pred_proba)
    }
        
    
    return metrics, y_pred, y_pred_proba, model
```

```python
baseline_results=[]
for model_name, model in models.items():
    print(f"\nTraining {model_name}...")
    try:
        metrics, y_pred, y_pred_proba, trained_model = evaluate_model(
            model, X_train_res, y_train_res, x_test_scaled, y_test, model_name
        )
        baseline_results.append(metrics)
        
        # Print metrics
        print(f"  Precision: {metrics['Precision']:.4f}")
        print(f"  Recall: {metrics['Recall']:.4f}")
        print(f"  F1-Score: {metrics['F1-Score']:.4f}")
        print(f"  ROC-AUC: {metrics['ROC-AUC']:.4f}")
        
    except Exception as e:
        print(f"  ✗ Error training {model_name}: {str(e)}")
```
