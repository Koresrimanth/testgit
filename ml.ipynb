```python
#fixing the outliers in the bill_amount using cqpping
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Define the Hybrid Capping + Log Function
def clean_billing_data(df, col):
    # Calculate IQR for the Upper Cap
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    # Define our "Hybrid" limits
    lower_limit = 0            # Set to 0 because negative debt is 0 risk
    upper_limit = Q3 + 1.5 * IQR # Standard IQR ceiling for outliers
    
    # Step A: Capping (Clip the extremes)
    df[col] = df[col].clip(lower=lower_limit, upper=upper_limit)
    
    # Step B: Log Transformation (Squash the distribution into a bell curve)
    # np.log1p handles the 0 values by calculating log(1 + x)
    df[col] = np.log1p(df[col])

# 2. Apply to all 6 Billing Columns
bill_cols = [f'BILL_AMT{i}' for i in range(1, 7)]

for col in bill_cols:
    clean_billing_data(df, col)

# 3. Visualize the Results
plt.figure(figsize=(15, 10))
for i, col in enumerate(bill_cols, 1):
    plt.subplot(2, 3, i)
    sns.histplot(df[col], kde=True, color='teal')
    plt.title(f'Cleaned & Logged {col}')
plt.tight_layout()
plt.show()

```
